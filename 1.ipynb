```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('s.jpg')
# Display the image
cv2.imshow('Loaded Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cropped_image = image[500:600, 600:800]

# Display the cropped image
cv2.imshow('Cropped Image', cropped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
'''100:300: This specifies the range of rows (height) to include in the crop.
100 is the starting row index (y-coordinate).
300 is the ending row index (exclusive).
200:400: This specifies the range of columns (width) to include in the crop.
200 is the starting column index (x-coordinate).
400 is the ending column index (exclusive).'''
```

```python
resized_image = cv2.resize(image, (500, 500))
# Display the resized image
cv2.imshow('Resized Image', resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
'''( new width , new height )'''
```

```python
#Convert to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)
'''128: This is the threshold value. Pixels with intensity values above this threshold 
will be set to the maximum value (255), while those below will be set to 0.
255: This is the maximum value to use with the THRESH_BINARY type.'''
'''_: This is a convention in Python to indicate that the first return value 
(which is the threshold used) is being ignored.'''
# Display the thresholded image
cv2.imshow('Binary Image', binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Finding Contours
contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
contour_image = image.copy()
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 3)
'''-1: This parameter specifies which contours to draw. Using -1 draws all contours.
If you want to draw a specific contour, you can provide its index.
(0, 255, 0): This tuple represents the color to draw the contours in BGR
format. In this case, it’s green.
3: This is the thickness of the contours being drawn'''

# Display the image with contours
cv2.imshow('Contours Image', contour_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Set Up Blob Detector Parameters
params = cv2.SimpleBlobDetector_Params()
#Create Blob Detector:
detector = cv2.SimpleBlobDetector_create(params)

# Detect blobs
keypoints = detector.detect(binary_image)

# Draw detected blobs as red circles
blob_image = cv2.drawKeypoints(image, keypoints, np.zeros((1, 1)), (0, 0, 255),
                               cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Display the image with blobs
cv2.imshow('Blobs Image', blob_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
